# Making changes to the data

```python
# Adding a new column call total
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 

print(df.head())

# Remove a column
df = df.drop(columns = ['Total'])
```

```python
# Accessing column via iloc[:,[begin,end)] the end is exclusive in numpy
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df
```

```python
# Rearranging column
cols = list(df.columns)
df2 = df[cols[0:4]+[cols[-1]]+cols[4:12]]
df2
```

# Write df to csv

```python
df2.to_csv("modified.csv",index = False) # By default ( index = True )it will write the row number
df2.to_excel("modified.xlsx",index = False) # By default ( index = True )it will write the row number
df2.to_csv("modified.txt",index = False,sep="\t") # By default ( index = True )it will write the row number
```

# Filtering data

```python
# Filter data with property of interest 

new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)] # Use & (and) ,| (or)
new_df
```

```python
# note that the old data frame index still remain, we can reset the index of new_df
new_df.reset_index(drop = True,inplace = True)
new_df
```

```python
# Remove Pokemon name with Mega
df.loc[df['Name'].str.contains('Mega')].head(5) # Filter only has Mega
```

```python
df.loc[~df['Name'].str.contains('Mega')].head(5) # ~ is not in pandas
```

```python
import re # regular expression 
df.loc[df['Type 1'].str.contains('Fire|Grass',regex = True)] 
```

```python
# re.I ignore upper/lower case
df.loc[df['Type 1'].str.contains('fiRe|grAss',flags = re.I,regex = True)]
```

```python
# some other patterns
df.loc[df['Name'].str.contains('^pi[a-z]',flags = re.I,regex = True)] # begin with ^
```

# Conditional Changes

```python
# Change type 1 grass to greener type
df.loc[df['Type 1'] == 'Grass','Type 1'] = "Greener"
df
```

```python
# Set all grass type to legendary pokemon
gdf = df.copy() # copy is required when correctly copy data.frame if not just address being assign
gdf.loc[gdf['Type 1']=='Greener','Legendary'] = True
gdf
```

```python
gdf.loc[gdf['Total']>500,['Generation','Legendary']] = ['Test 1','Test 2']
gdf
```

# Aggregate Statistics (Groupby)

```python
df['one'] = 1
df.groupby(['Type 1']).count()['one']
```

```python
df.groupby(['Type 1']).mean().sort_values('HP',ascending = False)
```

# Working with large amounts of data

```python
new_df2 = pd.DataFrame()
for dfx in pd.read_csv('pokemon_data.csv',chunksize = 5): # Load 5 row in a time
    dfx['one'] = 1
    results = dfx.groupby('Type 1').count()['one']
    new_df2 = pd.concat([new_df2,results])
    
# new_df2.groupby(0).sum()
new_df2.index.name = 'Type 1'
new_df2.reset_index(inplace = True)
new_df2.groupby('Type 1').sum()[0]
```
